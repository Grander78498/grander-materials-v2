```python
def rand(clusters, classes):
    numerator = 0
    denominator = 0
    for i in range(len(clusters)):
        for j in range(i + 1, len(clusters)):
            if (clusters[i] == clusters[j]) == (classes[i] == classes[j]):
                numerator += 1
            denominator += 1
    return numerator / denominator
```

```python
class DBSCAN:
    def __init__(self, 
                 data: pd.DataFrame, 
                 epsilon: float = 0.01, 
                 m: int = 10,
                 seed: int = 78498):
        self.X = data.values
        self.clusters = [-1 for _ in range(len(self.X))]
        self.clusters_count = 0
        self.visited = []

        self.epsilon = epsilon
        self.m = m

        np.random.seed(seed)

    def fit(self, x: pd.DataFrame | None = None):
        if x is not None:
            self.X = np.vstack(self.X, x)

        while len(self.visited) != len(self.X):
            index = np.random.randint(0, len(self.X))
            self._expand_cluster(index)


    def _expand_cluster(self, index: int, cluster: int | None = None):
        self.visited.append(index)
        x = self.X[index]
        point_neighbours = self._get_neighbours(x)
        if len(point_neighbours) >= self.m:
            if cluster is None:
                self.clusters_count += 1
                cluster = self.clusters_count
            
            for idx in point_neighbours:
                self._expand_cluster(idx, cluster)
        if cluster is not None:
            self.clusters[index] = cluster


    def _get_neighbours(self, x) -> dict[int, np.ndarray]:
        neighbours = dict()
        for idx, obj in enumerate(self.X):
            if 0.00001 < self._distance(x, obj) < self.epsilon:
                neighbours.update({idx: obj})
        return neighbours
    
    def _distance(x, y):
        return np.linalg.norm(x - y)
```
